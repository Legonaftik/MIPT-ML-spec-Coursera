```python
import numpy as np
from scipy.linalg import solve
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

f_1 = f(1)
f_4 = f(4)
f_8 = f(8)
f_10 = f(10)
f_15 = f(15)
first_args = np.array([[1, 1],
              [1, 15]])
first_vals = np.array([f_1, f_15])
first_coef = solve(first_args, first_vals)
print(first_coef)
first_x = np.array([1, 15])
first_y = first_coef[0] + first_coef[1] * first_x
plt.plot(first_x, first_y)
plt.show()
```

```python
second_args = np.array([[1, 1, 1],
                       [1, 8, 8**2],
                       [1, 15, 15**2]])
second_vals = np.array([f_1, f_8, f_15])
second_coef = solve(second_args, second_vals)
print(second_coef)
second_x = np.array([1, 8, 15])
second_y = second_coef[0] + second_coef[1]*second_x + second_coef[2]*(second_x**2)
plt.plot(second_x, second_y)
plt.show()
```

```python
third_args = np.array([[1, 1, 1, 1],
                     [1, 4, 4**2, 4**3],
                     [1, 10, 10**2, 10**3],
                     [1, 15, 15**2, 15**3]])
third_vals = np.array([f_1, f_4, f_10, f_15])
third_coef = solve(third_args, third_vals)
print(third_coef)
third_x = np.array([1, 4, 10, 15])
third_y = third_coef[0] + third_coef[1]*third_x + third_coef[2]*(third_x**2) + third_coef[3]*(third_x**3)
plt.plot(third_x, third_y)
plt.show()

with open("submission-2.txt", "w") as output:
    output.write(str(third_coef[0])+" "+str(third_coef[1])+" "+str(third_coef[2])+" "+str(third_coef[3]))
```
