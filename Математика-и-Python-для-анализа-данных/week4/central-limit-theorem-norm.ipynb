### Импортируем все необходимые нам в дальнейшем библиотеки.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

### Создаём выборку объёма 1000; строим гистограмму выборки и теоретическую плотность распределения поверх неё

```python
norm_rv = sts.norm(50, np.sqrt(10))
sample = norm_rv.rvs(1000)
plt.hist(sample, alpha=0.4, normed=True)

x = np.linspace(10, 90, 100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', color="red")
plt.legend()
plt.ylabel('frequency')
plt.xlabel('$x$')
```

### Сгенерируем 1000 выборок объёма 5 и построим гистограмму распределения их выборочных средних.


```python
sample_5 = [norm_rv.rvs(5) for i in range(1000)]
means_5 = [sum(n)/len(n) for n in sample_5]  # выборочные средние
plt.hist(means_5, alpha=0.4, normed=True)
plt.ylabel('frequency')
plt.xlabel('$x$')

norm_rv_5 = sts.norm(50, np.sqrt(10/5))  # здесь и в дальнейшем: EX=50, D=DX/n=10/n
x = np.linspace(35,65)
pdf = norm_rv_5.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', color="red")
plt.legend()
plt.ylabel('frequency')
plt.xlabel('$x$')
```

### Увеличим объём выборок до 10 и повторим операцию.

```python
sample_10 = [norm_rv.rvs(10) for i in range(1000)]
means_10 = [sum(n)/len(n) for n in sample_10]
plt.hist(means_10, alpha=0.4, normed=True)
plt.ylabel('frequency')
plt.xlabel('$x$')

norm_rv_10 = sts.norm(50, 10/10)  # EX=50, D=10/n
x = np.linspace(35, 65)
pdf = norm_rv_10.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', color="red")
plt.legend()
plt.ylabel('frequency')
plt.xlabel('$x$')
```

### Увеличим размер выборок до 50.

```python
sample_50 = [norm_rv.rvs(50) for i in range(1000)]
means_50 = [sum(n)/len(n) for n in sample_50]
plt.hist(means_50, alpha=0.4, normed=True)
plt.ylabel('frequency')
plt.xlabel('$x$')

norm_rv_50 = sts.norm(50, np.sqrt(10/50))  # EX=50, D=10/n
x = np.linspace(46, 54)
pdf = norm_rv_50.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', color="red")
plt.legend()
plt.ylabel('frequency')
plt.xlabel('$x$')
```

## Вывод:
### Практически полученные результаты на выборках довольно близки к теоретическим. По мере увеличения размера выборок точность тоже возрастает. 
### Так, при размере выборки равном 5 средние значения укладывались в диапазоне [35, 65]; когда же в выборках стало присутствовать по 50 значений, средние показатели уместились в отрезок [46, 54].
### Таким образом мы можем искать оптимальный размер выборок, при котором отчётливо видны общие тенденции, но результат при этом не слишком сильно сглажен.
